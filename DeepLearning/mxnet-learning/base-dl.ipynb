```python
%matplotlib inline
from IPython import display
from matplotlib import pyplot as plt
from mxnet import nd, autograd
import random
```

```python
# 生成数据集
num_inputs = 2
num_examples = 1000
true_w = [2,-3.4]
true_b = 4.2
features = nd.random.normal(scale=1,shape=(num_examples,num_inputs))
labels = true_w[0] * features[:,0] + true_w[1] * features[:,1] + true_b
labels += nd.random.normal(scale=0.01, shape=labels.shape)
```

```python
features[0], labels[0]
```

```python
def use_svg_display():
    display.set_matplotlib_formats('svg')

def set_figsize(figsize=(3.5,2.5)):
    use_svg_display()
    plt.rcParams['figure.figsize'] = figsize
    
set_figsize()
plt.scatter(features[:,1].asnumpy(), labels.asnumpy(),1); # 加分好只显示图
```

```python
def data_iter(batch_size, features, labels):
    num_examples = len(features)
    indices = list(range(num_examples))
    random.shuffle(indices) # 样本的读取顺序的随机
    for i in range(0, num_examples, batch_size):
        j = nd.array(indices[i: min(i+ batch_size, num_examples)])
        yield features.take(j), labels.take(j) # take函数根据索引返回对应元素
```

```python
batch_size = 10
for x,y in data_iter(batch_size, features, labels):
    print(x,y)
    break
```

```python
# 初始化 模型参数
w = nd.random.normal(scale=0.01, shape=(num_inputs,1))
b = nd.zeros(shape=(1,))
```

```python
w.attach_grad()
b.attach_grad()
```

```python
# 定义模型
def linreg(X,w,b):
    return nd.dot(X,w) +b 
```

```python
def squared_loss(y_hat, y):   # 定义的损失函数
    return (y_hat - y.reshape(y_hat.shape)) ** 2/2
```

```python
def sgd(params, lr,batch_size): # 定义优化算法
    for param in params:
        param[:] = param - lr * param.grad / batch_size  
```

```python
lr = 0.03
num_epochs = 3
net = linreg
loss = squared_loss

for epoch in range(num_epochs): # 训练模型num_epochs个迭代周期
    # 每次迭代周期中， 会使用训练数据集中所有的样本
    # 和y分别是最小批量样本的特征和标签
    for X,y in data_iter(batch_size, features, labels):
        with autograd.record():
            l = loss(net(X,w,b),y)  # l 是有关最小批量 x和y 的损失
        l.backward()
        sgd([w,b], lr, batch_size) # 使用最小批量随机梯度下降
    train_l = loss(net(features,w,b), labels)
    print('epoch %d, loss %f' % (epoch * i, train_l.mean().asnumpy())) 
```
